```python
def Outlier1(df2):
    threshold=3
    for i in range(len(df2.columns)):
        X=df2.iloc[:,i]
        outlier=[]
        
        mean_data=np.mean(X)
        std_data=np.std(X)
        for j in X:
            z_score=(j-mean_data)/std_data
            if np.abs(z_score)>threshold:
                outlier.append(j)
        print(outlier)
```

```python
Outlier1(df2)
```
